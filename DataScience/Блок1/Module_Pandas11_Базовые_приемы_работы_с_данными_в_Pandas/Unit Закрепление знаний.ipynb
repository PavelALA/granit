```python
citibyke_tripdata_df = citibyke_tripdata.copy()
citibyke_tripdata_df.head()
```

```python
citibyke_tripdata_df['age'] = 2018 - citibyke_tripdata_df['birth year']
citibyke_tripdata_df.drop('birth year',axis=1,inplace=True)
```

```python
citibyke_tripdata_df
```

```python
citibyke_tripdata_df[citibyke_tripdata_df['age']>60].shape[0]
```

```python
citibyke_tripdata_df['starttime'] = pd.to_datetime(citibyke_tripdata_df['starttime'])
citibyke_tripdata_df['stoptime'] = pd.to_datetime(citibyke_tripdata_df['stoptime'])
citibyke_tripdata_df
```

```python
citibyke_tripdata_df['trip duraion'] = citibyke_tripdata_df['stoptime'] - citibyke_tripdata_df['starttime']
seconds = citibyke_tripdata_df['trip duraion'].mean().seconds
seconds
```

```python
weekends_days = citibyke_tripdata_df['starttime'].dt.dayofweek
weekends_days
```

```python
citibyke_tripdata_df['weekend'] = weekends_days.apply(lambda x: 1 if x == 5 or x == 6 else 0 )
citibyke_tripdata_df
```

```python
citibyke_tripdata_df['weekend'].sum()
```

```python
day_time = citibyke_tripdata_df['starttime'].dt.hour
day_time
```

```python
def day_times(hour):
    if hour >= 0 and hour < 7:
        return 'night'
    elif hour > 6 and hour < 13:
        return 'morning'
    elif hour > 12 and hour < 19:
        return 'day'
    elif hour > 18 and hour < 24:
        return 'evening'
citibyke_tripdata_df['time_of_day'] = day_time.apply(day_times)
citibyke_tripdata_df
```

```python
display(citibyke_tripdata_df['time_of_day'].value_counts())
```
