```python
import pandas as pd
ratings_movies = pd.read_csv('data/ratings_movies.csv')
display(ratings_movies.tail())
```

```python
#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg)
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
display(ratings_movies.info())
```

```python
ratings_movies[ratings_movies['year_release']==1999].groupby('title')['rating'].agg(['mean']).sort_values(by='mean')
```

```python
ratings_movies[ratings_movies['year_release']==2010].groupby('genres')['rating'].agg(['mean']).sort_values(by='mean')
```

```python
ratings_movies.groupby('userId')['genres'].nunique().sort_values(ascending=False)
```

```python
ratings_movies.groupby('userId')['rating'].agg(['count','mean']).sort_values(by=['count','mean'],ascending=[True,False])
```

```python
ratings_movies[ratings_movies['year_release']==2018].groupby('genres')['rating'].agg(['count','mean']).sort_values(by=['count','mean'],ascending=[True,False])
```

```python
ratings_movies['year_rating'] = pd.to_datetime(ratings_movies['date']).dt.year
display(ratings_movies)
```

```python
ratings_movies.pivot_table(
    values='rating',
    index='year_rating',
    columns='genres',
    aggfunc='mean'
).sort_values(by='Comedy')['Comedy']
```
